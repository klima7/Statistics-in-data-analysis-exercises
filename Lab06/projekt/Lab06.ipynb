```python
from scipy.io.arff import loadarff
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
pd.options.mode.chained_assignment = None

df = pd.read_csv('../baza_danych/letter-recognition.data')
features = df.columns[1:]
df
```

Podział zbioru i normalizacja 
---

```python
R = 0.2
V = 0.1
train, test = train_test_split(df, test_size=R)


mms = MinMaxScaler()
mms.fit(train[features], train['Class'])
train[features] = mms.transform(train[features])
test[features] = mms.transform(test[features])
train, validate = train_test_split(train, test_size=V)
```

Selekcja cech
---

```python
from sklearn.feature_selection import SelectKBest

skb = SelectKBest(k = 12)

skb.fit(train[features], train['Class'])
selected_features = features[skb.get_support(indices=True)]

train_features = train[selected_features]
test_features = test[selected_features]
validate_features = validate[selected_features]
```

Klasyfikacja i ewaluacja
---

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import sklearn.metrics as metrics

def show_conf(y_true: np.ndarray, y_pred: np.ndarray, names: tuple):
    conf = metrics.confusion_matrix(y_true, y_pred, labels=list(names))
    print()
    print(list(names))
    print(y_true.unique())
    print(np.unique(y_pred))
    df_cm = pd.DataFrame(conf, index = list(names), columns = list(names))
    plt.figure(figsize = (30,8))
    sn.heatmap(df_cm, annot=True, fmt='g') 
    plt.show()

def evaluate(y_true, y_pred):
    acc = metrics.accuracy_score(y_true, y_pred)
    prec = metrics.recall_score(y_true, y_pred, average='macro')
    jacc = metrics.jaccard_score(y_true, y_pred, average='macro')
    print('Accuracy:', acc)
    print('Recall:', prec)
    print('Jaccard:', jacc)
```

Wnioski
---
- Normalizacja cech za pomocą `MinMaxScaler` pomogła uzyskać lepsze wyniki.
- Najlepsze wyniki zostały uzyskane dla wybranych za pomocą `SelectKBest` 12 cech z 16 istniejących. 
- Najlepsze wyniki dla `KNN` (`k` > 1) zostały uzyskane dla `k` równego 3.
- Największe wartości `Accuracy`, `Recall`, `Jaccard Score` uzyskuje klasyfikator `NN` (Nearest Neighbors), minimalnie gorsze uzyskuje `KNN` dla `k` równego 3, a najgorzej wypada `NM`.
